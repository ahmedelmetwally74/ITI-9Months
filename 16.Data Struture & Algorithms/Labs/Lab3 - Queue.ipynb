#### Implement a queue using array (size 5) for integer  the end will get a menue:
#### ==> when you choose number,it perform requierd task


1) add
2) delete
3) search
4) insert
5) display all
6) exit

```python
class Node:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.next = None

class QueueList:
    def __init__(self):
        self.head = None
        self.tail = None

    def enqueue(self, name, id):
        new_node = Node(name, id)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def dequeue(self):
        if self.head is not None:
            removed_node = self.head
            self.head = self.head.next
            if self.head is None:
                self.tail = None
            return removed_node
        else:
            return None

    def display_all(self):
        current = self.head
        while current:
            print(f"ID: {current.id}, Name: {current.name}")
            current = current.next

menu = """
1) Add
2) Delete
3) Search
4) Display All
5) Exit
"""

queue_list = QueueList()

while True:
    print(menu)
    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter ID to add: "))
        name = input("Enter Name: ")
        queue_list.enqueue(name, id)
        print(f"ID: {id}, Name: {name} added successfully to the queue.")

    elif choice == 2:
        removed_node = queue_list.dequeue()
        if removed_node:
            print(f"ID: {removed_node.id}, Name: {removed_node.name} removed successfully from the queue.")
        else:
            print("Queue is empty. Cannot dequeue.")

    elif choice == 3:
        print("Search operation not supported for a queue.")

    elif choice == 4:
        print("All Records in the Queue:")
        queue_list.display_all()

    elif choice == 5:
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Please enter a valid option.")
```
