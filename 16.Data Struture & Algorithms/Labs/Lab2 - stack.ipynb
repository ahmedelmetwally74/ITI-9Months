### implement a stack using linkedlist for student having (id,name) the end will get a menu :
### ==> when you choose number,it perform requierd task

1) add
2) delete
3) search
4) display all
5) exit

```python
class node:
    def __init__(self, name, id):
        self.name= name
        self.id = id 
        self.pre = None

class stack:
    def __init__(self):
        self.tail=None

    def push(self, name, id):
        nd = node(name,id)
        # Empty Stack
        if(self.tail==None):
            self.tail = nd
        else:
            nd.pre= self.tail
            self.tail=nd

    def pop(self):
        nd=self.tail 
        if(self.tail != None):
            nd = self.tail
            self.tail=nd.pre
        return nd 
    
    def display_all(self):
        cur = self.tail
        while cur:
            print(f"ID: {cur.id}, Name: {cur.name}")
            cur = cur.pre

menu = """
1) Add
2) Delete
3) Search
4) Display All
5) Exit
"""

student_stack = stack()

while True:
    print(menu)
    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter Student ID: "))
        name = input("Enter Name: ")
        student_stack.push(name, id)
        print(f"Student with ID {id} added successfully.")

    elif choice == 2:
        popped_node = student_stack.pop()
        if popped_node:
            print(f"Student with ID {popped_node.id} and Name {popped_node.name} deleted successfully.")
        else:
            print("Stack is empty. Cannot delete.")

    elif choice == 3:
        id_to_search = int(input("Enter Student ID to search: "))
        found = False
        current = student_stack.tail
        while current:
            if current.id == id_to_search:
                found = True
                print(f"Student found - ID: {current.id}, Name: {current.name}")
                break
            current = current.pre

        if not found:
            print(f"Student with ID {id_to_search} not found.")

    elif choice == 4:
        print("All Student Records in the Stack:")
        student_stack.display_all()

    elif choice == 5:
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Please enter a valid option.")

        
```
