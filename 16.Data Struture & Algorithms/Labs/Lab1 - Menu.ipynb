```python
class node:
    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.next = None
        self.pre = None

    def __str__(self):
        return f"ID: {self.id}, Name: {self.name}"


class linkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def add(self, id, name):
        new_node = node(id, name)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.pre = self.tail
            self.tail = new_node

    def insert(self, id, name, loc):
        new_node = node(id, name)

        if self.head is None:
            self.head = new_node
            self.tail = new_node
        elif loc == 0:
            new_node.next = self.head
            self.head.pre = new_node
            self.head = new_node
        else:
            i = 0
            cur = self.head

            while i < loc - 1 and cur is not None:
                cur = cur.next
                i += 1

            if cur == self.tail or cur is None:
                self.tail.next = new_node
                new_node.pre = self.tail
                self.tail = new_node
            else:
                cur.next.pre = new_node
                new_node.next = cur.next
                cur.next = new_node
                new_node.pre = cur

    def search(self, name):
        nd = self.head
        while nd is not None and nd.name != name:
            nd = nd.next
        return nd

    def delete_Modified(self, id):
        cur = self.head
        while cur is not None:
            if cur.id == id:
                while cur is not None and cur.id == id:
                    if cur == self.head:
                        self.head = self.head.next
                        if self.head:
                            self.head.pre = None
                        else:
                            self.tail = None
                    elif cur == self.tail:
                        self.tail = self.tail.pre
                        self.tail.next = None
                    else:
                        cur.pre.next = cur.next
                        cur.next.pre = cur.pre

                    cur = cur.next
            else:
                cur = cur.next

    def displayAll(self):
        cur = self.head
        while cur:
            print(cur)
            cur = cur.next


menu = """
1) Add
2) Delete
3) Search
4) Insert
5) Display All
6) Exit
"""

doubly_linked_list = linkedList()

while True:
    print(menu)
    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter Student ID: "))
        name = input("Enter Name: ")
        doubly_linked_list.add(id, name)

    elif choice == 2:
        id = int(input("Enter Student ID to delete: "))
        doubly_linked_list.delete_Modified(id)

    elif choice == 3:
        name = input("Enter Name to search: ")
        result = doubly_linked_list.search(name)
        if result:
            print(f"Student ID for {name}: {result.id}")
        else:
            print(f"Student with name {name} not found.")

    elif choice == 4:
        loc = int(input("Enter position to insert: "))
        id = int(input("Enter Student ID: "))
        name = input("Enter Name: ")
        doubly_linked_list.insert(id, name, loc)

    elif choice == 5:
        print("All Student Records:")
        doubly_linked_list.displayAll()

    elif choice == 6:
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Please enter a valid option.")
```
