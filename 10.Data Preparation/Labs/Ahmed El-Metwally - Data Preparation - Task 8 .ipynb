```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 
import requests 
```

```python
# Here I make a driver to work 
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
URL = 'https://www.jumia.com.eg/catalog/?q=mobile+samsung'
driver.get(URL)
```

```python
titles=driver.find_elements(By.XPATH,"//h3[@class='name']")
for title in titles:
    print(title.text)
```

```python
prices=driver.find_elements(By.XPATH,"//div[@class='prc']")
for price in prices:
    print(price.text)
```

```python
images=driver.find_elements("xpath","//img[@class='img']")
images_urls = [img.get_attribute('src') for img in images] 
print(len(images_urls))

for url in images_urls:
    print(url)
```

```python
# Download the images 
index = 0
for index, url in enumerate(images_urls):
    response = requests.get(url,stream=True)
    
    with open(f'img-{index+1}.jpg' ,'wb')as f:
        for chunk in response.iter_content(chunk_size=1000):
            print(chunk)
            f.write(chunk)
```
